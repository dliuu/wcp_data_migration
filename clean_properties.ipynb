```python
import pandas as pd
import numpy as np

pd.options.display.max_rows = None
pd.options.display.max_columns = None
```

```python
raw_df = pd.read_csv('raw_data/properties.csv', on_bad_lines='skip')
raw_df.info()
print(raw_df.head(10))
```

```python
df = raw_df.dropna(thresh=3000,axis=1, how='all')
only_na = raw_df[~raw_df.index.isin(df.index)]

print('Dropped Columns: \n')
print(only_na.columns)
```

```python
print('Finalized Columns: \n')
print(df.columns)
```

```python

df['wcp_bathroomsprop'].replace(0.0, np.nan, inplace=True)
df['wcp_bedroomsprop'].replace(0.0, np.nan, inplace=True)
df['wcp_sqftprop'].replace(0.0, np.nan, inplace=True)
df.head(10)

#nexrm_address1 = address line 1
#nexrm_city = city
#nexrm_state = state
#wcp_propertytypename = property type
#wcp_bedroomsprop = bedrooms
#nextrm_bathroomsprop = bathrooms
#wcp_sqftprop = sqft
#wcp_countyname = county
#nextrm_validatedaddress = full address

#nexrm_latitude
#nexrm_longitude

```
