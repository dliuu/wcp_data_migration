```python
import pandas as pd
import numpy as np

pd.options.display.max_rows = None
pd.options.display.max_columns = None
```

```python
raw_df = pd.read_csv('raw_data/loans.csv', on_bad_lines='skip')
raw_df.info()
raw_df.columns
raw_df.head(10)
```

```python
df = raw_df.dropna(thresh=3000,axis=1, how='all')
df.columns
df.info()
df.tail(10)

```

```python


convert_to_ints = ['nexrm_borrowercashreserves', 'nexrm_borrowercashreserves_base', 'nexrm_borrowerestimateafterrepair', 'nexrm_borrowerestimateasis', 
                   'nexrm_borrowerestimaterepaircosts', 'nexrm_borrowermoneyintotransaction', 'nexrm_closingfee', 'nexrm_constructiondraw', 'nexrm_downpayment',
                   'nexrm_purchaseprice', 'nexrm_valuationfee', 'nexrm_finalvalueasis', 'nexrm_finalvalueasrepaired', 'nexrm_finalrepairamount',
                   'wcp_wcploanamount', 'wcp_titleinsurance', 'wcp_netwire', 'wcp_commitmentfee', 'wcp_loanpayments', 'wcp_payofffundsamt', 
                   'wcp_increasenet', 'wcp_decreasenet', 'nexrm_loanagentcommission', 'nexrm_loanreferralcommission', 'nexrm_amounttoaddpayoff', 
                   'wcp_prepaidinterestamount', 'nexrm_prepaidinterestpaid', 'wcp_initialdraw', 'wcp_arvappraisalvalue', 'wcp_asisappraisalvalue',
                    'wcp_retailconstructcost', 'wcp_constructiondrawdisbursedamount', 'wcp_constructiondrawpendingamount', 'wcp_feestotalamountdue',
                    'wcp_assignmentfee', 'wcp_repairappraisal', 'wcp_totalprincipalpaiddown', 'wcp_currentloanamount', 'wcp_constructionfundbalance',
                    'wcp_estimatedrevenue', 'wcp_estimatedrevenueadjusted', 'wcp_prepaidinterestbalance', 'wcp_totalconstructionbudget', 'wcp_currentmaxloanfunding',
                    'wcp_currentfundingamount', 'wcp_hudinterest', 'wcp_perdiem', 'wcp_estimatedpayoffamount', 'wcp_accruedinterest', 'wcp_colchistotalcostbasis',
                    'wcp_rehabspendtodate', 'wcp_currentdrawnbalance']


df[convert_to_ints].dropna(axis=0, inplace=True)
df[convert_to_ints] = df[convert_to_ints].astype('str')
df[convert_to_ints].dtypes

def remove_chars(s):
    if isinstance (s, str):
        s = s.replace('$', '').replace(',', '').replace('<NA>', '')
        if s.startswith('.'):
            s = s[1:]
        if s.strip() == '':
            return np.nan
    return s

df[convert_to_ints] = df[convert_to_ints].applymap(lambda x: remove_chars(x) if isinstance(x, str) else x)

df[convert_to_ints].replace('<NA>', np.nan)
df[convert_to_ints].replace('', np.nan)
df[convert_to_ints].dropna(axis=0, how='any', inplace=True)



df[convert_to_ints] = df[convert_to_ints].astype('float')
```

```python
return_col = []
for col in df.columns:
    if 'paid' in col:
        return_col.append(col)

print(return_col)
```

```python
df['createdon'] = pd.to_datetime(df['createdon'])
df['wcp_loanfunded'] = pd.to_datetime(df['wcp_loanfunded'])
df['wcp_totalprincipalpaiddown_date'] = pd.to_datetime(df['wcp_totalprincipalpaiddown_date'])

active_df = df[(df['createdon'] >= '2023-01-01')]
active_df.append(df[(df['wcp_loanfunded'] >= '2022-01-01')])
active_df.append(df[(df['wcp_totalprincipalpaiddown_date'] >= '2023-01-01')])
active_df.head(5)
active_df.info()
```
