```python
import pandas as pd
import numpy as np

pd.options.display.max_rows = None
pd.options.display.max_columns = None

borrow_df = pd.read_csv('raw_data/borrowingentites.csv', on_bad_lines='skip')
borrow_df.info()
print(borrow_df.head(10))

title_df = pd.read_csv('raw_data/titlecompanies.csv', on_bad_lines='skip')
title_df.info()
```

```python
drop_threshold = len(borrow_df) * 0.5

df = borrow_df.dropna(thresh=drop_threshold,axis=1, how='all')
only_na = borrow_df[~borrow_df.index.isin(df.index)]

print('Dropped Columns: \n')
print(only_na.columns)
```

```python
print('Finalized Columns: \n')
print(df.columns)
```

```python
title_final = title_df.dropna(thresh=drop_threshold,axis=1, how='all')
only_na = title_df[~title_df.index.isin(title_final.index)]

print('Dropped Columns, Title Companies: \n')
print(only_na.columns)
```

```python
print('Finalized Columns: \n')
print(title_final.columns)
```
